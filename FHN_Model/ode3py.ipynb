# Ejemplo de la funcion ODE3 en python

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Definir la función de las ecuaciones diferenciales
def myODE(t, y):
    dydt = -2 * t * y
    return dydt

# Condiciones iniciales y tiempo de integración
y0 = [1.0]  # Condiciones iniciales
t_span = (0, 2)  # Tiempo de integración desde t=0 hasta t=2

# Resolver el sistema de ODEs
sol = solve_ivp(myODE, t_span, y0, method='RK45')

# Graficar la solución
plt.figure(figsize=(8, 6))
plt.plot(sol.t, sol.y[0], 'b', label='y(t)')
plt.xlabel('Tiempo t')
plt.ylabel('y')
plt.title('Solución de ODE usando solve_ivp')
plt.grid(True)
plt.legend()
plt.show()
```
